```python
import spacy
from cefrpy import CEFRSpaCyAnalyzer, CEFRLevel

nlp = spacy.load("en_core_web_sm")
text = "concierge"
text_analyzer = CEFRSpaCyAnalyzer(
    entity_types_to_skip={"PERSON", "ORG", "GPE"},
    abbreviation_mapping={"n't": "not", "'s": "is"},
)
tokens = text_analyzer.analize_doc(nlp(text))
for word, pos, skipped, level, *_ in tokens:
    if not skipped and level:
        print(word, pos, f"{level:.2f}", CEFRLevel(round(level)))
```
